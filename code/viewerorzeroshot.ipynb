```python
import json
import matplotlib.pyplot as plt
import matplotlib
from collections import defaultdict

output_file = '../data/cadec/output.json'
gt_file = '../data/cadec/test.json'
train_file = '../data/cadec/train.json'
with open(output_file,'r') as fp:
    output = json.load(fp)
with open(gt_file, 'r') as fp:
    gt = json.load(fp)
with open(train_file, 'r') as fp:
    train = json.load(fp)
```

```python
train[0]['ner']
```

```python
# try to analyze whether the model fails when there are new vacabs
train_vocab = set()
train_entity_vocab = set()
for sentence_idx in range(len(train)):
     sentence = train[sentence_idx]['sentence']
     ners = train[sentence_idx]['ner']
     for ner in ners:
          tokens = ' '.join([sentence[i] for i in ner['index']])
          train_entity_vocab.add(tokens)
     for token in sentence:
          train_vocab.add(token)
```

```python
train_entity_vocab
```

```python
'fog' in train_vocab
```

```python
#zero-shot-analysis
# the entity is never seen in train set
zero_shot_total = 0
zero_shot_correct = 0

for sentence_idx in range(len(gt)):
    sentence = gt[sentence_idx]['sentence']
    ners = gt[sentence_idx]['ner']

    for ner in ners:
        tokens = [sentence[i] for i in ner['index']]
        tokens = ' '.join(tokens)
        if tokens not in train_entity_vocab:

            zero_shot_total += 1
            tokens = [sentence[i] for i in ner['index']]
            tokens_type = {'text':tokens, 'type':'adr'}
            if tokens_type in output[sentence_idx]['entity']:
                zero_shot_correct += 1

print("the zero shot entity number:", zero_shot_total, "the precision on zero-shot:", zero_shot_correct/zero_shot_total)


# new vocab analysis:
# the entity in the test set has new vocabs
new_vocab_total = 0
new_vocab_correct = 0

for sentence_idx in range(len(gt)):
    sentence = gt[sentence_idx]['sentence']
    ners = gt[sentence_idx]['ner']

    for ner in ners:
        tokens = [sentence[i] for i in ner['index']]
        new_vocab = True
        for token in tokens:
            if token in train_vocab:
                new_vocab = False
                break

        if new_vocab:
            new_vocab_total += 1
            tokens = [sentence[i] for i in ner['index']]
            tokens_type = {'text':tokens, 'type':'adr'}
            if tokens_type in output[sentence_idx]['entity']:
                new_vocab_correct += 1

print("the new vocab entity number:", new_vocab_total, "the precision on new vocab:", new_vocab_correct/new_vocab_total)
```
